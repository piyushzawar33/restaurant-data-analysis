```python
users_df = pd.read_json('users.json')
users_df.head()
```

```python
from sqlite3 import connect
conn = connect(':memory:')
cursor = conn.cursor()

with open('restaurants.sql', 'r') as f:
    sql_script = f.read()

cursor.executescript(sql_script)
conn.commit()

restaurants_df = pd.read_sql_query('select * from restaurants', conn)
restaurants_df.head()
```

```python
orders_users_df = orders_df.merge(users_df, on="user_id", how="left")
orders_users_df.head()
```

```python
df = orders_users_df.merge(restaurants_df, on="restaurant_id", how="left", suffixes=("_order", "_restaurant"))
df.head()
```

```python
# Which city has the highest total revenue (total_amount) from Gold members?  

gold_df = df[df["membership"] == "Gold"]

city_revenue = gold_df.groupby("city", as_index=False)["total_amount"].sum().sort_values("total_amount", ascending=False)

city_revenue
```

```python
# Which cuisine has the highest average order value across all orders?
avg_order_value = df.groupby("cuisine", as_index=False)["total_amount"].mean().sort_values("total_amount", ascending=False)
avg_order_value
```

```python
# How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?

# Total spend per user
user_total_spend = df.groupby("user_id")["total_amount"].sum()

# Count of users with total spend > 1000
high_value_users_count = len(user_total_spend[user_total_spend > 1000])
high_value_users_count
```

```python
# Which restaurant rating range generated the highest total revenue?

rating_bins = [3.0, 3.5, 4.0, 4.5, 5.0]
rating_labels = ["3.0–3.5", "3.6–4.0", "4.1–4.5", "4.6–5.0"]

df["rating_range"] = pd.cut(df["rating"],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)

rating_revenue = df.groupby("rating_range", as_index=False)["total_amount"].sum().sort_values("total_amount", ascending=False)
rating_revenue
```

```python
# Among Gold members, city with highest average order value
gold_city_aov = df[df["membership"] == "Gold"].groupby("city", as_index=False)["total_amount"].mean().sort_values("total_amount", ascending=False)
gold_city_aov
```

```python
# What percentage of total orders were placed by Gold members? (Rounded to nearest integer)

percentage_gold_orders = round((df[df["membership"] == "Gold"].shape[0] / df.shape[0]) * 100)
percentage_gold_orders
```

```python
# Which restaurant has the highest average order value but less than 20 total orders?

# Filter required restaurants
filtered_df = df[df["restaurant_name_order"].isin([
    "Grand Cafe Punjabi",
    "Grand Restaurant South Indian",
    "Ruchi Mess Multicuisine",
    "Ruchi Foods Chinese"
])]

filtered_df
```

```python
# Count orders and average order value per restaurant
stats = filtered_df.groupby("restaurant_name_order").agg(
    avg_order_value=("total_amount", "mean"),
    total_orders=("order_id", "count")
)

stats
```

```python
stats = stats[stats["total_orders"] < 20]
stats
```

```python
# Which combination contributes the highest revenue?
df[
    (df["membership"].isin(["Gold", "Regular"])) &
    (
        ((df["membership"] == "Gold") & (df["cuisine"].isin(["Indian", "Italian"]))) |
        ((df["membership"] == "Regular") & (df["cuisine"].isin(["Indian", "Chinese"])))
    )
].groupby(["membership", "cuisine"])["total_amount"].sum().sort_values(ascending=False)
```

```python
# During which quarter of the year is the total revenue highest?
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

df.assign(
    order_date=pd.to_datetime(df["order_date"]),
    quarter=df["order_date"].dt.to_period("Q").astype(str)
).groupby("quarter")["total_amount"].sum().idxmax()
```
